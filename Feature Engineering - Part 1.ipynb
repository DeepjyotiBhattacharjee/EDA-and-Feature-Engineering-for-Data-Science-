# Train test split

```python
from sklearn.model_selection import train_test_split
```

```python
df
```

```python
X_train,X_test,y_train,y_test = train_test_split(df.drop("Purchased",axis=1),df["Purchased"],test_size=0.25)
```

```python
X_train
```

```python
X_test
```

# Standard Scaler

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scaler.fit(X_train) # we use fit to perform z transform only on the train data and fit_transform on test data to avoid data leakage.
```

fit calculates the mean and std dev and transform transforms the data. we use the same mean and std dev of the training data on our test data as well to avoid data leakage.

```python
X_train_scaled = scaler.transform(X_train)
X_train_scaled
```

```python
X_train
```

```python
X_test_scaled = scaler.transform(X_test)
X_test_scaled
```

```python
X_train_scaled = pd.DataFrame(X_train_scaled,columns=X_train.columns)
X_train_scaled
```

# Effect of scaling

```python
import matplotlib.pyplot as plt

fig,(ax1,ax2) = plt.subplots(ncols=2,figsize=(12,5))

ax1.scatter(X_train["Age"],X_train["EstimatedSalary"])
ax1.set_title("Before scaling")

ax2.scatter(X_train_scaled["Age"],X_train_scaled["EstimatedSalary"],color="red")
ax2.set_title("After scaling")
```
