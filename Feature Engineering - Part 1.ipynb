# Standard Scaler

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scaler.fit(X_train) # we use fit to perform z transform only on the train data and fit_transform on test data to avoid data leakage.
```

fit calculates the mean and std dev and transform transforms the data. we use the same mean and std dev of the training data on our test data as well to avoid data leakage.

```python
X_train_scaled = scaler.transform(X_train)
X_train_scaled
```

```python
X_train
```

```python
X_test_scaled = scaler.transform(X_test)
X_test_scaled
```

# When should we use min max and when should we use standardization ?

1. when we know the range of the data e.g., CGPA lies within 0 to 10 -> we use min max
2. when we don't know the range of the data, e.g., salary -> we use standard scaler

# Outlier Handling and Detection

```python
 # Now all the data outside of 3 standard deviations are outliers
```

```python
# upper limit
df["cgpa"].mean() + 3*df["cgpa"].std() # above this value is outlier
```

```python
# lower limit
df["cgpa"].mean() - 3*df["cgpa"].std() # below this value is outlier
```

```python
df[df["cgpa"]>8.80] # these are outliers
```

```python
df[df["cgpa"]<5.11] # these are outliers
```

```python
df[(df["cgpa"]>8.80) | (df["cgpa"]<5.11)] # these are the outliers in our data
```

## Capping

```python
upper_limit = df["cgpa"].mean() + 3*df["cgpa"].std()
lower_limit = df["cgpa"].mean() - 3*df["cgpa"].std()
```

```python
# wherever we find a value > upper limit -> replace it it upper limit
# wherever we find a value < lower limit -> replace it it lower limit
```

```python
df['cgpa']=np.where(df["cgpa"]>upper_limit,upper_limit,df["cgpa"])
```

```python
df['cgpa'].max()
```

```python
df['cgpa'].min()
```

```python
upper_limit
```

```python
lower_limit
```

```python
df['cgpa']=np.where(df["cgpa"]<lower_limit,lower_limit,df["cgpa"])
```
